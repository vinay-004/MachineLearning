```python
n_steps = 120
n_inputs = 21
n_neurons = 27
n_outputs = 27
batch_size = 1
```

```python
X = tf.placeholder(tf.float32, [batch_size,n_steps, n_inputs])
y = tf.placeholder(tf.float32, [batch_size,n_steps, n_outputs])
cell = tf.contrib.rnn.BasicRNNCell(num_units=n_neurons, activation=tf.nn.relu)
outputs, states = tf.nn.dynamic_rnn(cell, X, dtype=tf.float32)
```

```python
cell = tf.contrib.rnn.OutputProjectionWrapper(
tf.contrib.rnn.BasicRNNCell(num_units=n_neurons, activation=tf.nn.relu),
output_size=n_outputs)
```

```python
learning_rate = 0.001
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=outputs, labels=y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
n_iterations = 1000
```

```python
i=0
j=0
with tf.Session() as sess:
    init.run()
    for iteration in range(n_iterations):
        X_batch = np.reshape(x_train_3d[i,j:j+120],(batch_size,n_steps,n_inputs))
        y_batch = np.reshape(y_train_3d[i,j:j+120],(batch_size,n_steps,n_outputs))
        
        #X_batch, y_batch = [] # fetch the next training batch
        sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        if iteration % 100 == 0:
            mse = loss.eval(feed_dict={X: X_batch, y: y_batch})
            print(iteration, "\tMSE:", mse)
        
        if(j >= 47878):
            j == 0
            i=(i+1)%44
        else:
            j = j + 120
    y_pred = sess.run(outputs, feed_dict={X: np.reshape(x_test_3d[0,0:120],(batch_size,n_steps,n_inputs))})
    print(y_pred)
```
